```python
import nibabel as nb
import numpy as np
import scipy.io
import scipy
from scipy import stats
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from ipynb.fs.full.utils import *
```

```python
Mean_FPPs_Activity = mean_FPPs_movie_data()
```

```python
Mean_FPPs_Activity.shape
```

```python
data = Mean_FPPs_Activity

correlation_matrix = np.zeros((5, 5))
pvalue_matrix = np.zeros((5, 5))

for i in range(5):
    for j in range(5):
        correlation_matrix[i, j], pvalue_matrix[i, j] = pearsonr(data[i], data[j])

plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'Arial'

plt.figure(figsize=(10, 6))
plt.imshow(correlation_matrix, cmap='YlOrRd', interpolation='none')
cbar = plt.colorbar(label='Pearson correlation', pad=0.05)

cbar.set_label('Pearson correlation', labelpad=10) 
# plt.title('Correlation Matrix')
plt.xticks(np.arange(5), ['FPP1', 'FPP2', 'FPP3', 'FPP4', 'FPP5'])
plt.yticks(np.arange(5), ['FPP1', 'FPP2', 'FPP3', 'FPP4', 'FPP5'])

# plt.savefig('E:/FPP_files_and_codes/Figures/Figure4_files/Row1/correlation_matrix.png', dpi=600, bbox_inches='tight')
plt.show()
```
